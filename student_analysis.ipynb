# Data Analysis

```python
# Calculate the average final grade
average_grade = df["G3"].mean()

print("Average Final Grade:", round(average_grade, 2))
```

### Observation

The average final grade of all students is approximately **10.42**.This indicates that the overall academic performance of the students is moderate, with most students scoring around the middle of the grading scale.

```python
# Students scoring above 15
high_score_students = df[df["G3"] > 15]

print("Number of students scoring above 15 :", len(high_score_students))
```

### Observation

Out of 395 students, **40 students** scored more than **15 marks**, indicating that only a small percentage of students achieved excellent academic performance.

```python
# Correlation of numerical features with final grade (G3)
correlation = df.corr(numeric_only=True)

correlation["G3"].sort_values(ascending=False)
```

## Observation

The second-period grade (**G2**) has the strongest positive correlation (**0.9049**) with the final grade (**G3**), followed by the first-period grade (**G1**) with a correlation of **0.8015**. This indicates that students who perform well throughout the academic year are likely to achieve higher final grades.

Among the negative correlations, the number of past failures (**-0.3604**) has the strongest negative relationship with the final grade, suggesting that students with more previous failures generally score lower in the final examination.

```python
# Average final grade by gender
gender_performance = df.groupby("sex")["G3"].mean()

print(gender_performance)
```

# Data Visualization

```python
import matplotlib.pyplot as plt

gender_performance.plot(kind="bar")

plt.title("Average Final Grade by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.show()
```

## Observation

The average final grade of male students is **10.91**, while female students have an average final grade of **9.97**. Based on this dataset, male students performed slightly better on average than female students.

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10, edgecolor="black")

plt.title("Distribution of Final Grades")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")

plt.show()
```

## Observation

The histogram shows the distribution of students' final grades. Most students scored in the middle range, while only a few students achieved very high or very low grades.

```python
plt.figure(figsize=(8,5))

plt.scatter(df["studytime"], df["G3"])

plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")

plt.show()
```

## Observation

The scatter plot suggests a weak positive relationship between study time and final grades. Although students who study more often tend to score better, study time alone does not fully determine academic performance.

```python
travel_performance = df.groupby("traveltime")["G3"].mean()

travel_performance.plot(kind="bar")

plt.title("Travel Time vs Average Final Grade")
plt.xlabel("Travel Time")
plt.ylabel("Average Final Grade")

plt.show()
```

## Observation

Students with shorter travel times generally have slightly higher average grades. As travel time increases, average academic performance tends to decrease slightly.
